```python
import pandas as pd
import seaborn as sns
df = pd.read_csv('events.csv')
df_event_conversion = df.loc[df['event']=='conversion', :]
df_event_conversion_cols = df_event_conversion[['timestamp', 'person', 'model']]
df_event_conversion_cols.columns = ['FechaCompra', 'person', 'model']
df_event_conversion_cols['visita'] = 1
#df_event_conversion_cols.iloc[0:30]
df_event_conversion_cols_grouped = df_event_conversion_cols.groupby(['person', 'model']).agg({'visita':['sum'], 'FechaCompra':['min']})
df_event_conversion_cols_grouped.iloc[0:30]
```

```python
df_event_viewed = df.loc[(df['event']=='viewed product'), :]
df_event_viewed_cols = df_event_viewed[['timestamp', 'person', 'model']]
df_event_viewed_cols.columns = ['FechaVisita', 'person', 'model']
df_event_viewed_cols_grouped = df_event_viewed_cols.groupby(['person', 'model']).agg({'FechaVisita':['min']})
df_event_viewed_cols_grouped.iloc[0:30]
```

```python
df_join = pd.merge(df_event_conversion_cols_grouped, df_event_viewed_cols_grouped, how='inner', on=['person', 'model']).reset_index()
df_join.columns = ['person', 'model', 'visita', 'FechaCompra', 'FechaVisita']
df_final = df_join.loc[df_join['FechaCompra'] > df_join['FechaVisita']]

#df_final.sort_values(by='visita', ascending=[False]).iloc[0:30]
```

```python
#df_final = df_join.loc[df_join['FechaCompra'] > df_join['FechaVisita']]
#df_final.columns = ['person', 'model', 'visita', 'FechaCompra', 'FechaVisita']
df_final_count = df_final.groupby(['visita']).count().reset_index()
df_final_count = df_final_count[['visita', 'person']]
df_final_count.iloc[0:40]


#ax = sns.boxplot(x = df_final2['visita']['sum'])


#df_lead_conversion_cols = df_lead_conversion[['timestamp','event', 'person', 'model']]
#df_lead_conversion_cols.head()
```

```python
colores = ['#4d3dff', '#ffad48', '#63ff73', '#d92b45', '#db15a3', '#696764', '#737421']

p = sns.countplot(data=df_final, x = 'visita',edgecolor=(0,0,0), linewidth=2, palette=colores )
p.axes.set_yscale('log')
p.set_title('Cantidad de visitas del celular previas a su compra')
p.set_ylabel('cantidad (log)')
p.set_xlabel('Visitas')
p.grid(True, which='major', linestyle='-.', axis='x' )
p.grid(True, which='major', linestyle='-.', axis='y' )
for patch in p.patches:
    x=patch.get_bbox().get_points()[:,0]
    y=patch.get_bbox().get_points()[1,1]
    p.annotate('{:.0f}'.format(y), (x.mean(), y), 
            ha='center', va='bottom') # set the alignment of the text
```
