```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
datos = pd.read_csv('events.csv')
datos
```

```python
regiones = datos[["region", "person"]].dropna().drop_duplicates("person")
regiones = regiones.groupby("region")
regiones = regiones.count().reset_index().sort_values("person", ascending=False)
regiones = regiones.head(30)
regiones
```

```python
regiones["person in log"] = np.log(regiones["person"]+1)
```

```python
ax = sns.barplot(x=regiones['region'], y=regiones['person in log'],edgecolor=(0,0,0), linewidth=2)
ax.set_title('Las primeras 30 regiones con mayor cantidad de gente que entra a la web',fontsize=20)
ax.set_ylabel('Cantidad de personas (log)',fontsize=14)
ax.set_xlabel('Regiones',fontsize=14)
ax.grid(True, which='major', linestyle='-.', axis='y' )
ax.set_xticklabels(regiones['region'],rotation=90)
ax.figure.set_figheight(6)
ax.figure.set_figwidth(6)
```
