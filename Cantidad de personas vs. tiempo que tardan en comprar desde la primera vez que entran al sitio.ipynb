```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
datos = pd.read_csv('events.csv')
datos
```

```python
datos_gente_compro = datos[datos["event"] == "conversion"].drop_duplicates("person")
datos_gente_compro = datos_gente_compro[["timestamp", "person"]]

#Renombro las columnas para que queden en la tabla final luego de hacer un join. 
datos_gente_compro.columns = ["timestamp2", "person"]
datos_gente_compro
```

```python
#Ordeno por fecha y me quedo con la primera aparición de cada persona.
datos_primera_accion = datos.sort_values("timestamp")
datos_primera_accion = datos[["timestamp", "person"]].drop_duplicates("person")
datos_primera_accion
```

```python
#Hago un join de las dos tablas para quedarme con las personas que compraron, 
#la fecha en la cual entraron por primera vez y la fecha en la cual compraron.

datos_gente = pd.merge(datos_gente_compro, datos_primera_accion, on = "person", how = "inner")
```

```python
#Resto las fechas (timestamp2 - timestamp)
datos_gente['timestamp'] = pd.to_datetime(datos_gente['timestamp'])
datos_gente['timestamp2'] = pd.to_datetime(datos_gente['timestamp2'])

datos_gente["tiempo en comprar"] = datos_gente["timestamp2"] - datos_gente["timestamp"]
```

```python
datos_10_bins = datos_gente.groupby(pd.cut(datos_gente["tiempo en comprar"], 10, labels = ('(0 - 13)', '(13 - 26)', '(26 - 40)','(40 - 53)','(53 - 67)','(67 - 80)','(80 - 94)', '(94 - 107)', '(107 - 120)', '(120 - 134)'))).count()
```

```python
datos_10_bins = datos_10_bins[["person"]]
datos_10_bins.reset_index()
```

```python
x = np.logspace(0, 5, num=21)
y = (np.sin(1.e-2*(x[:-1]-20))+3)**10

p = datos_10_bins.plot(kind='bar',    # Plot a bar chart
        legend=False,    # Turn the Legend off
        width=0.75,      # Set bar width as 75% of space available
        figsize=(8,5.8),  # Set size of plot in inches
        color=[plt.cm.Paired(np.arange(len(datos_10_bins)))],
        logy=True)

p.set_xlabel("Tiempo que tardan en comprar (en días)", size = 16)
p.set_ylabel("Cantidad de personas", size = 16)
p.set_title("Cantidad de personas vs. tiempo que tardan en comprar desde la primera vez que entran al sitio", size = 22)

#x = range(0, 135)
#plt.stackplot(x, datos_gente, labels=['A','B','C', 'D', 'E', 'F', 'G', 'H', 'I', 'J'])
```
