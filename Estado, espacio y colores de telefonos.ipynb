```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tabla_completa = pd.read_csv('events.csv');
tabla_completa
```

```python
tabla_completa['event']=tabla_completa['event'].astype('category')
```

```python
conversion_r = tabla_completa[tabla_completa['event']=='conversion']
conversion_r.head()
```

```python
conversion = conversion_r.drop(["url","skus","search_term","staticpage","campaign_source","search_engine","channel","new_vs_returning","city","region","country","device_type","screen_resolution","operating_system_version","browser_version"],axis='columns')
conversion.head()
```

```python
condition = conversion_r.groupby('condition').agg({'person':'count'})
condition.sort_values(ascending=False,by="person")
condition.reset_index()
```

```python
cell = sns.countplot(data=conversion, x = 'condition',edgecolor=(0,0,0), linewidth=2, order = conversion['condition'].value_counts().index )
cell.set_title('Condicion de los celulares')
cell.grid(True, which='major', linestyle='-.', axis='y' )

   
for i in cell.patches:
    # get_x pulls left or right; get_height pushes up or down
    cell.text(i.get_x()+.15, i.get_height()+.5, \
            str(round((i.get_height()), 2)), fontsize=13,
                color='black')
```

```python
colores = ['#000000', '#FFD700', '#ffffff', '#3d3b3b', '#C0C0C0', '#b76e79', '#EE99AC', '#808080', '#7038F8', '#8B0000' ]
cell = sns.countplot(data=conversion, y = 'color',edgecolor=(0,0,0), linewidth=2, order = conversion['color'].value_counts().iloc[0:10].index, palette = colores )
cell.set_title('Top 10 de colores de celulares m√°s comprados')
cell.grid(True, which='major', linestyle='-.', axis='x' )

for i in cell.patches:
    cell.text(i.get_width()+.1, i.get_y()+.50, \
            str(round((i.get_width()), 2)), fontsize=13, color='black')
```

```python
#['4GB','8GB','16GB','32GB','64GB','128GB','256GB','512GB']
cell = sns.countplot(data=conversion, x = 'storage',edgecolor=(0,0,0), linewidth=2, order = conversion['storage'].value_counts().index )
cell.set_title('Espacio de los celulares')
cell.grid(True, which='major', linestyle='-.', axis='y' )

   
for i in cell.patches:
    # get_x pulls left or right; get_height pushes up or down
    cell.text(i.get_x()+.15, i.get_height()+.5, \
            str(round((i.get_height()), 2)), fontsize=13,
                color='black')
```
