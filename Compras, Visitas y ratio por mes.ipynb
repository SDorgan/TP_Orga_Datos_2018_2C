```python
import pandas as pd
import seaborn as sns; sns.set()
import matplotlib.pyplot as plt
df = pd.read_csv('events.csv')
df
```

```python
df = df.sort_values(by='timestamp')
df_filter_event_person = df[['event','person']]#.iloc[0:20]
df_filter_event_person['compras'] = df_filter_event_person['event'] == 'conversion'
#df_filter_person.iloc[0:20]
df_filter_person_action = df_filter_event_person[['person', 'compras']]
#bla.groupby(['person']).sum()
df_filter_timestamp = pd.to_datetime(df['timestamp']).dt.month
#df_filter_timestamp = df_filter_timestamp['timestamp'].apply(lambda x:x.strftime('%Y%m'))
df_filter_date_person_action = df_filter_timestamp.to_frame().join(df_filter_person_action)
#df_filter_date_person_action.iloc[0:20]
df_group = df_filter_date_person_action.groupby(['timestamp', 'person']).sum().reset_index()
df_group['visitas'] = True
#df_group.iloc[0:20]
df_date_buy_visit = df_group[['timestamp','compras', 'visitas']]
df_final = df_date_buy_visit.groupby(['timestamp']).sum().reset_index()
#df_group[['compras']].sum()
#df_final.iloc[0:50]
df_final['ratio'] = df_final.apply(lambda row: row.compras / row.visitas, axis=1)
df_final[['timestamp','compras', 'visitas', 'ratio']].plot.line(subplots=True, x="timestamp", figsize=(10,10))
#df_final[['compras']].sum()
```
