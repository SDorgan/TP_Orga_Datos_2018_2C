# Modos Por Los Que Se Ingresa Al Sitio
El siguiente grfico muestra los distintos canales por los se ingres al sitio de Trocafone en el eje X y en el eje Y la cantidad de ingresos que hubo por ese canal.

```python
p = sns.countplot(data=vs_no_unknown, x = 'channel',edgecolor=(0,0,0), linewidth=2, order = vs_no_unknown['channel'].value_counts().index)
p.axes.set_yscale('log')
p.set_title('Modos Por Los Que Se Ingresa Al Sitio')
p.set_ylabel('cantidad (log)')
p.set_xlabel('Canal de Ingreso')
p.grid(True, which='major', linestyle='-.', axis='x' )
p.grid(True, which='major', linestyle='-.', axis='y' )
for patch in p.patches:
    x=patch.get_bbox().get_points()[:,0]
    y=patch.get_bbox().get_points()[1,1]
    p.annotate('{:.0f}'.format(y), (x.mean(), y), 
            ha='center', va='bottom') # set the alignment of the text
```
