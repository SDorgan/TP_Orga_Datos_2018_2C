```python
import pandas as pd
import seaborn as sb
import seaborn as sns
```

```python
datos = pd.read_csv('events.csv')
```

```python
datos_event = datos[datos["event"] == "ad campaign hit"]
datos_ = datos[["timestamp","person","event","campaign_source"]].dropna()
```

```python
datos_['timestamp'] = pd.to_datetime(datos_['timestamp'])
datos_['hour'] = datos_['timestamp'].dt.hour
datos_['weekday_name'] = datos_['timestamp'].dt.weekday_name

weekdays = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]

cross = pd.crosstab(datos_.hour, datos_.weekday_name)
cross = cross.reindex(columns=weekdays)
cross.head()

g = sb.heatmap(cross,  cmap="BuPu")
g.set_title("Cantidad de ad campaign hits por hora y día de semana", fontsize=22)
g.set_xlabel("Día de semana",fontsize=16)
g.set_ylabel("Horario", fontsize=16)
```

```python
datos_['month'] = datos_['timestamp'].dt.month

cross = pd.crosstab(datos_.month, datos_.weekday_name)
cross = cross.reindex(columns=weekdays)
cross.head()

g = sb.heatmap(cross,  cmap="BuPu")
g.set_title("Cantidad de ad campaign hits por mes y día de semana", fontsize=22)
g.set_xlabel("Día de semana",fontsize=16)
g.set_ylabel("Mes", fontsize=16)
```
